```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("Order ID is not a numeric column, so mean cannot be calculated.")
print("Number of unique Order IDs:", df["Order ID"].nunique())
```

```python
print("Mean of Amount",df["Amount"].mean())
print("Median of Amount",df["Amount"].median())
print("Mode of Amount", df["Amount"].mode())
```

```python
total_revenue = df['Amount'].sum()
total_profit = df['Profit'].sum()
total_quantity = df['Quantity'].sum()
avg_order_value = df['Amount'].mean()
profit_margin = (total_profit / total_revenue) * 100

print(f"Total Revenue: ₹{total_revenue:,}")
print(f"Total Profit: ₹{total_profit:,}")
print(f"Total Items Sold: {total_quantity:,}")
print(f"Average Order Value (AOV): ₹{avg_order_value:.2f}")
print(f"Overall Profit Margin: {profit_margin:.2f}%")
```

```python
sns.histplot(df['Amount'])
```

```python
df.groupby('Category')[['Amount', 'Profit']].sum().sort_values(by='Profit', ascending=False)
```

```python
category_data = df.groupby('Category')[['Amount', 'Profit']].sum().reset_index()
sns.barplot(x='Category', y='Profit', data=category_data)
plt.title('Profit by Category')
plt.show()
```
